```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("book_data.csv")
df_books = df[0:6]
df_series = df[6:12]
df_updated = df[12:18]
```

```python
def plot_type(df, title, xlabel, ylabel, filename):
    plt.style.use('ggplot')
    x = np.array(df["Title"])
    y = np.array(df["Sales"])
    plt.xticks(rotation=70)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(filename, bbox_inches='tight')
```

```python
plot_type(df_books, "Book sales vs Book", "Book", "Sales (100 million)", "books.png")
```

```python
plot_type(df_series, "Series sales vs Book series", "Series", "Sales (100 million)", "series.png")
```

```python
plot_type(df_updated, "Regularly-updated book sales vs Book", "Book", "Sales (10 million)", "updated.png")
```
